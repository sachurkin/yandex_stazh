```python
import pandas as pd
import numpy as np
import datetime
```

Загрузка и выделение нужных данных

```python
data = pd.read_csv('Reklamny_analitik_stazhirovka.csv', sep = ';')
data = data.drop(columns = 'client_id')
data.groupby('Category').count()
```

```python
data['Month'] = np.vectorize(lambda x: datetime.datetime.strptime(x, "%d.%m.%Y"))(data['Month'])
```

```python
data = data[(data['Month'] < datetime.datetime.strptime('01.09.2020', "%d.%m.%Y")) & (data['Month'] >= datetime.datetime.strptime('01.01.2020', "%d.%m.%Y"))]
```

```python
data = data[data['Category'] == 'ПОТРЕБИТЕЛЬСКОЕ КРЕДИТОВАНИЕ']
```

```python
data
```

```python
data = data[(data['Shows'] != 0)]
data
```

```python
data['Cost_rub_wo_NDS'] = np.vectorize(lambda x: float(x.replace(',','.')))(data['Cost_rub_wo_NDS'])
```

Подсчет эффективности затрат всех клиентов в общем

```python
X = data.groupby(['Client']).sum()
X['Эффективность затрат'] = X['Сonversions']/X['Cost_rub_wo_NDS']*1000
X
```

```python
ax = X['Эффективность затрат'].plot.bar(x='lab', y='val', rot=0)
```

```python
X['Показы/1000 рублей'] = X['Shows']/X['Cost_rub_wo_NDS']*1000
ax = X['Показы/1000 рублей'].plot.bar(x='lab', y='val', rot=0)
```

```python
X['Переходы/1000 рублей'] = X['Clicks']/X['Cost_rub_wo_NDS']*1000
ax = X['Переходы/1000 рублей'].plot.bar(x='lab', y='val', rot=0)
```

Подсчет эффективности затрат разных типов рекламы, использованных нашим клиентом

```python
Z = data[data['Client'] == 'Омега'].groupby(['Device','Place', 'BannerType', 'QueryType (search only)','TargetingType']).sum()
Z['Эффективность затрат'] = Z['Сonversions']/Z['Cost_rub_wo_NDS']*1000
Z
```

Подсчет эффективности затрат разных типов рекламы, для всех клиентов в среднем

```python
Y = data.groupby(['Device','Place', 'BannerType', 'QueryType (search only)','TargetingType']).sum()
Y['Эффективность затрат'] = Y['Сonversions']/Y['Cost_rub_wo_NDS']*1000
Y.sort_values(by=['Эффективность затрат'], ascending = False)
```
